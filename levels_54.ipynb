```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('levels_54.csv', index_col = 'full_level_index')
f, ax = plt.subplots(1,2, figsize = (10,4))
sns.rugplot(y = df.height_m, height = 0.5, ax=ax[0])
sns.rugplot(y = df.height_m[df.height_m < 1000], height = 0.5, ax=ax[1])
```
